##**Hierarchical clustering (Agglomerative Hierarchical clustering)**

**Method** 

* First we define each point as cluster i.e we have N data points then we N cluster gets created.
* Find the closest pair of cluster and merge them in one cluster.
* Now we keep doing that until we get one cluster of size N. 
* But suppose you stop at during the way , you produce a dendogram.
* It kind of greedy approach , because we keep making locally optimal decision at each point.

there's agglomerative clustering or heirarchial clustering where nothing is assumed in terms of number of clusters

there, you start with each datapoint as its own centroid ( population as 1) and you group points till all points grouped into 1 big cluster ( population is N)
so where you break this tree will give those many clusters .

* The basic idea behind `k-means` consists of defining k clusters such that total within-cluster variation (or error) is minimum.

* A cluster center is the representative of its cluster. The squared distance between each point and its cluster center is the required variation. The aim of k-means clustering is to find these k clusters and their centers while reducing the total error.

## **These methods are**:

* The Elbow Method
* The Silhouette Method

##**Elbow method**

Calculate the Within-Cluster-Sum of Squared Errors (WSS) for different values of k, and choose the k for which WSS becomes first starts to diminish. In the plot of WSS-versus-k, this is visible as an elbow.

[<img src="https://miro.medium.com/max/814/1*O_JmBi6rM6PlrPFx_uNrVQ.png"/>](https://miro.medium.com/max/814/1*O_JmBi6rM6PlrPFx_uNrVQ.png)

As expected, the plot looks like an arm with a clear elbow at **`k = 3`**.<br/>

Unfortunately, we do not always have such clearly clustered data. This means that the elbow may not be clear and sharp.

## **The Silhouette Method**

The silhouette value measures how similar a point is to its own cluster (cohesion) compared to other clusters (separation).<br/>

The range of the Silhouette value is between +1 and -1. A high value is desirable and indicates that the point is placed in the correct cluster. If many points have a negative Silhouette value, it may indicate that we have created too many or too few clusters.

The Silhouette Value s(i) for each data point i is defined as follows:

[<img src="https://miro.medium.com/max/590/1*vFXfyIjIjI_kbZNWODiuxQ.jpeg"/>](https://miro.medium.com/max/590/1*vFXfyIjIjI_kbZNWODiuxQ.jpeg) <br/>

### **`Note`**:

s(i) is defined to be equal to zero if i is the only point in the cluster. This is to prevent the number of clusters from increasing significantly with many single-point clusters.<br/>
Here, a(i) is the measure of similarity of the point i to its own cluster. It is measured as the average distance of i from other points in the cluster.

[<img src="https://miro.medium.com/max/806/1*OFOXVdMIwldl8diJvscw_A.jpeg"/>](https://miro.medium.com/max/806/1*OFOXVdMIwldl8diJvscw_A.jpeg) <br/>

Similarly, b(i) is the measure of dissimilarity of i from points in other clusters.

[<img src="https://miro.medium.com/max/586/1*fGgqT3IHwpEOiDbCtN5dAQ.jpeg"/>](https://miro.medium.com/max/586/1*fGgqT3IHwpEOiDbCtN5dAQ.jpeg) <br/>

**`d(i, j)`** is the distance between points i and j. Generally, Euclidean Distance is used as the distance metric.

```
## import package

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```
# generate artificial data
from sklearn.datasets import make_blobs
```

```
# n_sample : total number of points equally divided among clusters i.e. number of sample per cluster
# n_feature : number of feature of each sample
# centers : number of centers to the blob
# cluster_std : Standard deviation of the clusters
# random_state : state whould remain same as re-run the code again (that means randomly chooses data would remain same as we re-run the code) 
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```
data[0].shape
```

```
## visualize data

plt.scatter(data[0][:,0],data[0][:,1])  
plt.show()
```

```
# As center is 4 then we 4 sets of groups

plt.scatter(data[0][:,0],data[0][:,1], c = data[1] , cmap="rainbow")
plt.show()
```

```
from sklearn.cluster import KMeans
```

```
## Haha :) we already know that the number of cluster = 4 
## But on real data we don't know how many no. of cluster should we take 
kmeans = KMeans(n_clusters=4)
```

```
kmeans.fit(data[0])
```

```
## 4 Cluster center
kmeans.cluster_centers_
```

```
### these are labels that we created
data[1]
```

```
len(set(data[1]))
```

```
## count of actual label 
import numpy as np

values, counts = np.unique(data[1], return_counts=True)

print("Values : {} and counts : {}".format(values,counts))
```

```
## labels that predicted by kmeans
kmeans.labels_
```

```
## count of K-means label 
import numpy as np

values, counts = np.unique(kmeans.labels_, return_counts=True)

print("Values : {} and counts : {}".format(values,counts))
```

```
## visualize

fig,(ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(10,6))   ## sharey = True means they share same axis

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1], c = kmeans.labels_, cmap="rainbow")


ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1], c = data[1], cmap="rainbow")
plt.show()
```
